```python
miss = ['Mlle','Mme','Miss']
mrs = ['Ms','Lady','the Countess','Dr','Mrs']
mr = ['Dr','Major','Capt','Sir','Don','Mr']
others = ['Jonkheer','Col','Rev']
```

```python
data['Title_2'] = data.Title
data.loc[data.Title.isin (miss),'Title_2']='Miss'
data.loc[(data.Title.isin (mrs)) & (data.Sex == 'female'),'Title_2' ]='Mrs'
data.loc[(data.Title.isin (mr))& (data.Sex == 'male'),'Title_2']='Mr'
data.loc[data.Title.isin (others),'Title_2']='Other'
```
